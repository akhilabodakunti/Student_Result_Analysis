# Drop unnamed column

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

# Change weekly Study hours column

```python
df["WklyStudyHours "] = df["WklyStudyHours"].str.replace("5-10","6-10")
df.head()
```

```python
df["ParentMaritalStatus"] = df["ParentMaritalStatus"].str.replace("unmarried","married")
df.head()
```

# Gender Distribution

```python
plt.figure(figsize = (4,3))
ax=sns.countplot(data = df,x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Disributon")
plt.show()
```

```python
# from the above chart we have analysed that:
# the number of females in the data is morethan the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot = True)
plt.title("RelationShip between Parents Education and students Score")
plt.show()
```

```python
# from the above the chart we have concluded that the education of the parents have a good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot = True)
plt.title("RelationShip between Parents Marital Status And Students Score")
plt.show()
```

```python
# from the above chart we have concluded that there is no / negligible impact on the students score due to their parents marital status.
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
print(groupA["EthnicGroup"])
l = ["group A","group B","group C","group D","group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist,labels = l,autopct = "%1.2f%%")
plt.title("Distribution of EthnicGroups")
#plt.pie([])
plt.show()
```
